Modifikasi model CNN pada praktikum 2 sehingga didapatkan akurasi testing lebih dari 80%.



```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Modifikasi CNN CIFAR10 (>80% Akurasi Test)



```python

(cifar_train_images, cifar_train_labels), (cifar_test_images, cifar_test_labels) = keras.datasets.cifar10.load_data()
cifar_train_images = cifar_train_images.astype('float32')/255.0
cifar_test_images  = cifar_test_images.astype('float32')/255.0
class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
print('Train:', cifar_train_images.shape, 'Test:', cifar_test_images.shape)

augment = keras.Sequential([
    layers.RandomFlip('horizontal'),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1)
], name='augment')
BATCH_SIZE=64
train_ds = tf.data.Dataset.from_tensor_slices((cifar_train_images, cifar_train_labels))\
           .shuffle(10000).batch(BATCH_SIZE).map(lambda x,y:(augment(x, training=True), y))\
           .prefetch(tf.data.AUTOTUNE)
val_ds = tf.data.Dataset.from_tensor_slices((cifar_test_images, cifar_test_labels))\
         .batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

def make_cifar_model():
    inputs = layers.Input(shape=(32,32,3))
    x = inputs
    
    for _ in range(2):
        x = layers.Conv2D(32,3,padding='same',activation='relu')(x)
        x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D()(x); x = layers.Dropout(0.25)(x)
   
    for _ in range(2):
        x = layers.Conv2D(64,3,padding='same',activation='relu')(x)
        x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D()(x); x = layers.Dropout(0.35)(x)
  
    for _ in range(2):
        x = layers.Conv2D(128,3,padding='same',activation='relu')(x)
        x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D()(x); x = layers.Dropout(0.40)(x)
    x = layers.Flatten()(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(10, activation='softmax')(x)
    return keras.Model(inputs, outputs)

model_cifar = make_cifar_model()
model_cifar.compile(optimizer=keras.optimizers.Adam(1e-3),
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])
model_cifar.summary()


callbacks=[
    keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=8, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, verbose=1)
]

history_cifar = model_cifar.fit(train_ds, epochs=50, validation_data=val_ds, callbacks=callbacks)


loss_cifar, acc_cifar = model_cifar.evaluate(cifar_test_images, cifar_test_labels, verbose=0)
print(f'Akurasi CIFAR10 Test: {acc_cifar:.4f}')

hist_df = pd.DataFrame(history_cifar.history)
plt.figure(figsize=(6,4))
plt.plot(hist_df['accuracy'], label='train_acc')
plt.plot(hist_df['val_accuracy'], label='val_acc')
plt.title('CIFAR10 Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Acc'); plt.legend(); plt.show()

if acc_cifar >= 0.80:
    print('Target >80% tercapai.')
else:
    print('Target belum tercapai, lakukan tuning (lr, dropout, filter, epoch).')
```

Model CNN MNIST (Klasifikasi Angka 0-9)


```python
from tensorflow import keras
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import pandas as pd
```

```python

(mnist_train_images, mnist_train_labels), (mnist_test_images, mnist_test_labels) = keras.datasets.mnist.load_data()
mnist_train_images = mnist_train_images.astype('float32')/255.0
mnist_test_images  = mnist_test_images.astype('float32')/255.0
mnist_train_images = mnist_train_images[..., None]
mnist_test_images  = mnist_test_images[..., None]
print('Train:', mnist_train_images.shape, 'Test:', mnist_test_images.shape)

def make_mnist_model():
    model = models.Sequential([
        layers.Conv2D(32,3,activation='relu',input_shape=(28,28,1)),
        layers.MaxPooling2D(),
        layers.Conv2D(64,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(10, activation='softmax')
    ])
    return model

mnist_model = make_mnist_model()
mnist_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
mnist_model.summary()

history_mnist = mnist_model.fit(mnist_train_images, mnist_train_labels,
                                epochs=10, batch_size=64,
                                validation_split=0.1, verbose=1)

loss_mnist, acc_mnist = mnist_model.evaluate(mnist_test_images, mnist_test_labels, verbose=0)
print(f'Akurasi MNIST Test: {acc_mnist:.4f}')

hist_mnist_df = pd.DataFrame(history_mnist.history)
plt.figure(figsize=(6,4))
plt.plot(hist_mnist_df['accuracy'], label='train_acc')
plt.plot(hist_mnist_df['val_accuracy'], label='val_acc')
plt.title('MNIST Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Acc'); plt.legend(); plt.show()
```
